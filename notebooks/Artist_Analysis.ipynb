```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_2023 = pd.read_csv("../data/data_2023_clean.csv")
data_2024 = pd.read_csv("../data/data_2024_clean.csv")
```

## Analysis on 2023

```python
data_2023_groupby_artist = data_2023.groupby("Artist") # make the groupby object
```

```python
# Number of songs in the 2023 list group by artists

plt.bar(data_2023_groupby_artist["Track"].count().sort_values(ascending=False).head(10).index, 
        data_2023_groupby_artist["Track"].count().sort_values(ascending=False).head(10).values)

plt.xlabel("Artist")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.title("Number of songs in the 2023 list group by artists")
```

```python
# Total Streams by artists in 2023

plt.bar(data_2023_groupby_artist["Streams 2023"].agg('sum').sort_values(ascending=False).head(10).index, 
        data_2023_groupby_artist["Streams 2023"].agg('sum').sort_values(ascending=False).head(10).values)

plt.xlabel("Artist")
plt.ylabel("Streams")
plt.xticks(rotation=90)
plt.title("Total Streams by artists in 2023")
```

```python
data_2023.columns
```

```python
# Total appearance on charts by artists in 2023

data_2023["Total Chart Count"] = data_2023["Apple Music Chart Counts 2023"] + data_2023["Spotify Chart Counts 2023"] + data_2023["Deezer Chart Counts 2023"] + data_2023["Shazam Chart Counts 2023"]
plt.bar(data_2023_groupby_artist["Total Chart Count"].agg('sum').sort_values(ascending=False).head(10).index, 
        data_2023_groupby_artist["Total Chart Count"].agg('sum').sort_values(ascending=False).head(10).values)

plt.xlabel("Artist")
plt.ylabel("Streams")
plt.xticks(rotation=90)
plt.title("Total appearance on charts by artists in 2023")
```

- From these bar charts, we can see that in 2023, Taylor Swift was the most preferred artists. She has the greatest number of songs in the list, is in the second rank in the table of number of appearances on charts, and has greatest number of streams. 

- The Weekend is also a prefered artists, who appear in all there charts and is in the highest rank in the chart of number of appearances in Spotify Chart

- Other prefered artists are Ed Sheeran, Olivia Rodrigo, Bad Bunny, SZA, ...

## Analysis on 2024

```python
data_2024_groupby_artist = data_2024.groupby("Artist") # make the groupby object
```

```python
# Number of tracks by artists in 2024

plt.bar(data_2024_groupby_artist["Track"].count().sort_values(ascending=False).head(10).index, 
        data_2024_groupby_artist["Track"].count().sort_values(ascending=False).head(10).values)

plt.xlabel("Artist")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.title("Number of tracks by artists in 2024")
```

- In 2024, Taylor Swift is still the artist that has the greatest number of tracks in the dataset. Bad Bunny, Billie Eilish and The Weekend are also still in the top 10

- Drake has a significant success in this year. In 2023, his number of selected songs were not in top 10, but this year, he has about 60 tracks in the selected list, which is only 1 fewer than Taylor Swift.

- There are some noticeable new names in the top 10 such as Travis Scott or Ariana Grande

```python
# Total Streams by artists in 2024

data_2024["Total Streams"] = data_2024["Spotify Streams 2024"] + data_2024["Pandora Streams 2024"] + data_2024["Soundcloud Streams 2024"]
plt.bar(data_2024_groupby_artist["Total Streams"].agg('sum').sort_values(ascending=False).head(10).index, 
        data_2024_groupby_artist["Total Streams"].agg('sum').sort_values(ascending=False).head(10).values)

plt.xlabel("Artist")
plt.ylabel("Streams")
plt.xticks(rotation=90)
plt.title("Total Streams by artists in 2024")
```

- This table also shows that Drake is having a successful year. He is in the highest rank in the list of artists who have the most number of streams on 3 platforms: Spotify, Pandora, Soundcloud

- The Weekend, Bad Bunny, Taylor Swift or Ed Sheeran still maintain their high rank in this ranking

- There are some new noticeable names such as Post Malone, Ariana Grande, Maroon 5 or Post Malone

- There is a completely new name from Japan: MUSIC LAB JPN.

```python
# Mean Score by artists (Top 20 most popular) in 2024

top_artist_having_most_songs_2024 = data_2024_groupby_artist["Track"].count().sort_values(ascending=False).head(20).index.to_list()
top_artist_having_most_songs_2024 = data_2024[data_2024["Artist"].isin(top_artist_having_most_songs_2024)]
top_artist_group_df = top_artist_having_most_songs_2024.groupby("Artist")

plt.bar(top_artist_group_df["Track Score 2024"].agg('mean').sort_values(ascending=False).index, 
        top_artist_group_df["Track Score 2024"].agg('mean').sort_values(ascending=False).values)

plt.xlabel("Artist")
plt.ylabel("Streams")
plt.xticks(rotation=90)
plt.title("Mean Score by artists (Top 20 most popular) in 2024")
```

- This table show the mean score of tracks of artists who are in the list of top 20 artist who have the greatest number of tracks in the dataset. From this table, Billie Eilish and Post Malone have the highest rank

- Taylor Swift, The Weekend or Bad Bunny have a lot of tracks in the list, and their average score are also really high. It shows that their tracks are really prefered this year

- Drake is not in the top 20. It suggests that some Drake's songs didn't have good score in some criteria.
